## Cropping Images

```python
mac.crop((0,0,100,100)) # 0,0 is the top left corner, so how far it extends right is distance from x
# how far it extends down is distance from y
```

So this image now extends 100 pixels right and 100 pixels down. <br />
**Note:** 0 (intial x), 0 (inital y), 100 (distance from x), 100 (distance from y)

## Color Transparency
### RGBA Red, Green, Blue, Alpha

If Alpha is zero image is completely transparent; if its 255 its completely opaque

```python
red = Image.open("red_color.jpg")
```

```python
#red.show()
```

```python
blue = Image.open("blue_color.png")
```

```python
#blue.show()
```

```python
blue
```

How to change Alpha

```python
blue = blue.convert("RGB") # Was getting an error before searching for this conversion
```

```python
blue.putalpha(66)
```

```python
blue
```

```python
red
```

```python
red.putalpha(99)
```

```python
red
```

```python
blue.paste(im=red, box=(0,0),mask = red)
```

```python
blue
```
