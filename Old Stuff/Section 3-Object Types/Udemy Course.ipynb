Data Types:
Integers - int - Whole numbers such as: 3, 20, 0
Floating point - float  - Numbers with decimal point: 2.3, .4, 100.0
Strings - str - Ordered sequences of characters: "hello", "John", "2000"
Lists - list - Ordered sequence of objects [10, 20, 30, "Skipper"]
Dictionaries - dict - Unordered Key:Value pairs: {'myKey': 'value', 'name':'Johnny'}
Tuples - tup - Ordered immutable sequence of objects (10, 'hello', 200.3)
Sets - set - Unordered collection of unique objects: {"a","b"}
booleans - bool - Logical value indicating Ture or False
